# DuckDB Query Notebook
Connect to the database produced by `ingest_csvs.py` and explore the data.

```python
# ── USER CONFIG ───────────────────────────────────────────────────────────────
DB_FILE   = "ingested.duckdb"   # path to your .duckdb file
TABLE     = "unified_data"      # table name set in ingest_csvs.py
# ─────────────────────────────────────────────────────────────────────────────
```

```python
import duckdb
import pandas as pd

con = duckdb.connect(DB_FILE, read_only=True)
print(f"Connected to: {DB_FILE}")
```

## 1 — Explore the schema

```python
# All tables in the database
con.execute("SHOW TABLES").df()
```

```python
# Column names and types for the main table
con.execute(f"DESCRIBE {TABLE}").df()
```

## 2 — Quick look at the data

```python
# Row count
con.execute(f"SELECT COUNT(*) AS total_rows FROM {TABLE}").df()
```

```python
# First 10 rows
con.execute(f"SELECT * FROM {TABLE} LIMIT 10").df()
```

```python
# Summary statistics for every column
con.execute(f"SUMMARIZE {TABLE}").df()
```

## 3 — Filter rows

```python
# ← edit the WHERE clause to match your actual column names and values
con.execute(f"""
    SELECT *
    FROM {TABLE}
    WHERE your_column = 'some_value'
    LIMIT 20
""").df()
```

## 4 — Aggregate

```python
# Count rows grouped by a column  ← replace 'your_column'
con.execute(f"""
    SELECT your_column, COUNT(*) AS n
    FROM {TABLE}
    GROUP BY your_column
    ORDER BY n DESC
""").df()
```

## 5 — Pull into a Pandas DataFrame for further work

```python
df = con.execute(f"SELECT * FROM {TABLE}").df()

# From here you can use df like any normal pandas DataFrame
df.info()
```

```python
df.describe(include="all")
```

## 6 — Write an arbitrary SQL query

```python
# Paste any SQL you like here
query = f"""
    SELECT *
    FROM {TABLE}
    LIMIT 5
"""

con.execute(query).df()
```

```python
# Always close the connection when done
con.close()
print("Connection closed.")
```
